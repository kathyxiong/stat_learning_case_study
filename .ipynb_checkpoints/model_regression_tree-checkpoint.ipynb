```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# read in student math performance data set
# use original encoding (numeric for binary and ordinal vars)
student = pd.read_csv('data/student/student-por.csv', sep = ';')
```

```python
student.head(10)
```

```python
y = student['G3']
X = student.drop('G3', axis=1)
[student.shape, y.shape, X.shape]
```

```python
# label encode ordinal vars
ord_vars = ['Medu', 'Fedu', 'traveltime', 'studytime', 'failures'] # we are reverting back to numeric encoding
student[ord_vars] = student[ord_vars].apply(LabelEncoder().fit_transform)
```

```python
# one-hot encode nominal vars
nom_vars = student.columns[student.dtypes == 'category']
nom_vars_df = student[nom_vars]

onehotenc = OneHotEncoder()

onehotenc_df = nom_vars_df.apply(onehotenc.fit_transform).toarray()

head(onehotenc_df)
```
